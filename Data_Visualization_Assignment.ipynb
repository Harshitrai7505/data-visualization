# Q5- Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
#heatmap.

To generate a synthetic dataset with correlated features and visualize the correlation matrix using a heatmap, you can use the following code:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility

# Create a base feature
n_samples = 100
feature_1 = np.random.normal(0, 1, n_samples)

# Generate correlated features
feature_2 = feature_1 + np.random.normal(0, 0.5, n_samples)  # Feature 2 is correlated with Feature 1
feature_3 = feature_1 * 0.5 + np.random.normal(0, 1, n_samples)  # Feature 3 is less correlated

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})

# Add title
plt.title('Correlation Matrix Heatmap')

# Show plot
plt.show()
```

### Explanation:
1. **Import Libraries**: Import `numpy`, `pandas`, `matplotlib.pyplot`, and `seaborn` for data handling and visualization.
2. **Generate Synthetic Dataset**:
   - Use `np.random.seed(42)` for reproducibility.
   - Create a base feature (`feature_1`) with 100 samples from a normal distribution.
   - Create `feature_2` as a noisy version of `feature_1` to ensure high correlation.
   - Create `feature_3` as a combination of `feature_1` with less correlation.
3. **Create DataFrame**: Combine the features into a Pandas DataFrame.
4. **Calculate Correlation Matrix**: Use the `corr` method to calculate the correlation coefficients between the features.
5. **Heatmap**: Use `seaborn` to create a heatmap of the correlation matrix, with annotations and a color scale.
6. **Title**: Add a title to the heatmap for clarity.
7. **Show Plot**: Finally, display the plot with `show`.

You can run this code in a Python environment with the necessary libraries installed to visualize the correlation matrix of the synthetic dataset.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.layouts import column

# Set output to notebook
output_notebook()

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create a Bokeh figure
p = figure(title="Heatmap Example",
           x_axis_label='X',
           y_axis_label='Y',
           width=600,
           height=600)

# Define the color mapping
mapper = linear_cmap(field_name='value', palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

# Create a data source for the heatmap
data_source = {
    'x': np.repeat(x, len(y)),
    'y': np.tile(y, len(x)),
    'value': data_heatmap.flatten()
}

# Add rectangles to the figure
p.rect(x='x', y='y', width=0.1, height=0.1, source=data_source,
        line_color=None, fill_color=mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
